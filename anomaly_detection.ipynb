```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
data_path = "/home/vsanil/datasets/Numenta_Anomaly_Benchmark"
task_type = "artificialWithAnomaly"
task_path = os.path.join(data_path, task_type)
```

```python
data_df = pd.DataFrame()
for idx, filename in enumerate(os.listdir(task_path)):
    filepath = os.path.join(task_path, filename)
    ts_df = pd.read_csv(filepath)

    if idx==0:
        data_df = ts_df
        data_df.rename(columns={'value':filename}, inplace=True)
    else:
        data_df[filename] = ts_df['value']
        
plt.figure()
data_df.plot(subplots=True, figsize=(20,20))
plt.legend(loc='best')
plt.show()
```
